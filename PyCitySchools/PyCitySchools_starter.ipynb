```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_group = school_data_complete.groupby(["school_name"])
students = school_group["student_name"].count()

budget = school_group["budget"].median()
student_budget = budget / students
avg_math = school_group["math_score"].mean()
avg_reading = school_group["reading_score"].mean()
passing_rate = (avg_math + avg_reading)/2
math_pass = school_data_complete.loc[school_data_complete["math_score"] >= 70].groupby("school_name")["math_score"].count()/students*100
reading_pass = school_data_complete.loc[school_data_complete["reading_score"] >= 70].groupby("school_name")["reading_score"].count()/students*100
passing_rate = (math_pass + reading_pass)/2

school_df = pd.DataFrame({"Total Students" : students,
                            "Total School Budget" : budget,
                            "Per Student Budget" : student_budget,
                            "Average Math Score" : avg_math,
                            "Average Reading Score" : avg_reading,
                            "% Passing Math" : math_pass,
                            "% Passing Reading" : reading_pass,
                            "% Overall Passing Rate" : passing_rate
    
})


```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
top_df = school_data[["school_name","type"]].merge(school_df, on="school_name").set_index("school_name").sort_values("% Overall Passing Rate", ascending=False)
top_df["Total Students"] = top_df["Total Students"].map("{:,}".format)
top_df["Total School Budget"] = top_df["Total School Budget"].map("${:,.2f}".format)
top_df["Per Student Budget"] = top_df["Per Student Budget"].map("${:,.2f}".format)
top_df["Average Math Score"] = top_df["Average Math Score"].map("{:.4f}%".format)
top_df["Average Reading Score"] = top_df["Average Reading Score"].map("{:.4f}%".format)
top_df["% Passing Math"] = top_df["% Passing Math"].map("{:.4f}%".format)
top_df["% Passing Reading"] = top_df["% Passing Reading"].map("{:.4f}%".format)
top_df["% Overall Passing Rate"] = top_df["% Overall Passing Rate"].map("{:.4f}%".format)
top_df.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
bottom_df = school_data[["school_name","type"]].merge(school_df, on="school_name").set_index("school_name").sort_values("% Overall Passing Rate", ascending=True)

bottom_df["Total Students"] = bottom_df["Total Students"].map("{:,}".format)
bottom_df["Total School Budget"] = bottom_df["Total School Budget"].map("${:,.2f}".format)
bottom_df["Per Student Budget"] = bottom_df["Per Student Budget"].map("${:,.2f}".format)
bottom_df["Average Math Score"] = bottom_df["Average Math Score"].map("{:.4f}%".format)
bottom_df["Average Reading Score"] = bottom_df["Average Reading Score"].map("{:.4f}%".format)
bottom_df["% Passing Math"] = bottom_df["% Passing Math"].map("{:.4f}%".format)
bottom_df["% Passing Reading"] = bottom_df["% Passing Reading"].map("{:.4f}%".format)
bottom_df["% Overall Passing Rate"] = bottom_df["% Overall Passing Rate"].map("{:.4f}%".format)
bottom_df.head()
```
